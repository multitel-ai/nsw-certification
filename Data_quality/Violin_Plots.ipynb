# Reading in the .csv data file

```python
data_dir = "/App/Data/split_flight"
data_all = {}


df1 = pd.read_csv(os.path.join(data_dir, "train.csv"), index_col='index')
df1['type'] = 'train'
df2 = pd.read_csv(os.path.join(data_dir, "test.csv"), index_col='index')
df2['type'] = 'test'
df3 = pd.read_csv("/App/Data/split_quad/lourd.csv", index_col="index")
df3 = df3.drop(columns=['power'])
df3['type'] = 'lourd'
df4 = pd.read_csv(os.path.join(data_dir, "da.csv"))
df4['type'] = 'da'

data_all = pd.concat([df1, df2, df3, df4])
```

```python
#fig, axes = plt.subplots(figsize=(5,5))

plt.figure(figsize=(15,10))
ax = sns.violinplot(x=data_all["power"], y=data_all["type"], palette='pastel')
ax = sns.violinplot(x=data_all["power_smoothed"], y=data_all["type"], palette='pastel')
plt.show()
```
