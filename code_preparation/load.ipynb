```python
def read_split_csv(folder_path):
    """Load csv files in folder path in a dictionnary of pandas dataframes"""
    datasets = dict()
    for f in os.listdir(folder_path):
        print(f)
        datasets[f[:-4]] = pd.read_csv(os.path.join(folder_path, f), index_col=0)
    return datasets
```
