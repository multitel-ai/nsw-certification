```python
import pandas as pd 
import os
```

PREPROCESSING : \
La puissance est lissée et les vols sont rognés (sans le debut et la fin où le drone ne vole pas). 

SPLIT : \
Les jeux de données sont melangés et split (20% sont prélevés pour le test et puis 10% sont prélevés pour la validation).


```python
def read_split_csv(folder_path):
    """Load csv files in folder path in a dictionnary of pandas dataframes"""
    datasets = dict()
    for f in os.listdir(folder_path):
        print(f)
        datasets[f[:-4]] = pd.read_csv(os.path.join(folder_path, f), index_col=0)
    return datasets
```
